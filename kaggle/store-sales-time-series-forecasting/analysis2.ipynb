```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
train = pd.read_csv('data/train_merged.csv', parse_dates=['date'])
test = pd.read_csv('data/test_merged.csv', parse_dates=['date'])
```

```python
store1 = train[train.store_nbr == 1]
store1.family.value_counts()
```

```python
automotive1 = store1[store1.family == 'AUTOMOTIVE'].set_index('date')
automotive1.sort_index(inplace=True)
automotive1.head()
```

```python
fig = sm.graphics.tsa.plot_acf(automotive1.sales, lags=730)
fig.set_size_inches((20, 15))
```
