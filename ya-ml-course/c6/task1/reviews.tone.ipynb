```python
import pandas as pd
import numpy as np
from nltk.corpus import movie_reviews
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from operator import itemgetter
```

```python
negids = movie_reviews.fileids('neg')
posids = movie_reviews.fileids('pos')
```

```python
negfeats = [movie_reviews.words(fileids=[f]) for f in negids]
posfeats = [movie_reviews.words(fileids=[f]) for f in posids]
```

```python
data = []
for words in negfeats:
    data.append(dict(
        text=' '.join(words),
        positive=0,
    ))

for words in posfeats:
    data.append(dict(
        text=' '.join(words),
        positive=1,
    ))

data = pd.DataFrame(data)
```

```python
data.head()
```

```python
data.shape
```

```python
with open('ans1.txt', 'w') as fp:
    fp.write(str(data.shape[0]))
```

```python
pos_prop = data[data.positive == 1].shape[0] / data.shape[0]
pos_prop
```

```python
with open('ans2.txt', 'w') as fp:
    fp.write(str(pos_prop))
```

```python
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(data.text)
features.shape
```

```python
with open('ans3.txt', 'w') as fp:
    fp.write(str(features.shape[1]))
```

```python
estimator = LogisticRegression()
acc_cv_result = cross_val_score(estimator, features, data.positive, scoring='accuracy')
acc_cv_result
```

```python
mean_acc = round(np.mean(acc_cv_result), 4)
mean_acc
```

```python
with open('ans4.txt', 'w') as fp:
    fp.write(str(mean_acc))
```

```python
estimator = LogisticRegression()
auc_cv_result = cross_val_score(estimator, features, data.positive, scoring='roc_auc')
auc_cv_result
```

```python
mean_auc = round(np.mean(auc_cv_result), 4)
mean_auc
```

```python
with open('ans5.txt', 'w') as fp:
    fp.write(str(mean_auc))
```

```python
estimator.fit(features, data.positive)
```

```python
estimator.coef_
```

```python
est_features = []
feature_names = vectorizer.get_feature_names_out()
for idx, x in enumerate(estimator.coef_[0]):
    est_features.append((x, abs(x), idx, feature_names[idx]))

est_features.sort(key=itemgetter(1), reverse=True)
est_features
```

```python
with open('ans6.txt', 'w') as fp:
    fp.write(f'{est_features[0][-1]} {est_features[1][-1]}')
```
